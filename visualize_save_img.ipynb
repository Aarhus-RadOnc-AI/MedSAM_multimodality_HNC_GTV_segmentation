# Visualize and save the .npy images


```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
import os
join = os.path.join
from PIL import Image

array_path = '/processing/jintao/medsam_hnc/npy/HNC_noT1/imgs'
png_path = '/processing/jintao/medsam_hnc/npy/HNC_noT1/pngs'

```

```python
file_name = 'HNCDL_001-033.npy'
img_path = join(array_path, file_name) 
img_3c = np.load(img_path)
```

```python
# %% visualize results
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

last_channel_img = img_3c[:, :, -1]
print(last_channel_img.shape)
ax[0].imshow(img_3c)
ax[0].set_title("image full")
ax[1].imshow(last_channel_img)
ax[1].set_title("image last channel")
plt.show()
```

```python
# %% visualize results
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

last_channel_img = img_3c[:, :, -1]
print(last_channel_img.shape)
ax[0].imshow(img_3c)
ax[0].set_title("image full")
ax[1].imshow(last_channel_img)
ax[1].set_title("image last channel")
plt.show()
```

```python
png_file_path = join(png_path, file_name.replace('npy','png'))
# Assuming img is your NumPy array representing the image
img = Image.fromarray(np.uint8(img_3c*255))
img.save(png_file_path)
```
